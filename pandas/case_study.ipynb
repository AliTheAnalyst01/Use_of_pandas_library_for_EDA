# 16-spliting the string into multiple columns 

```python
df = pd.DataFrame({'name':['syed faizan','ch ans','meesum khan','turab iqbal'],
                   'location':['Lahore,Pakistan','Okara,Pakistan','Karachi,Pakistan','Henburg,Germany']})
df
```

```python
df.name.str.split(' ',expand=True)
```

```python
df[['first_name','last_name']]=df.name.str.split(' ',expand=True)
```

```python
df
```

```python
df[['city','country']]=df.location.str.split(',',expand=True)
df
```

```python
df = df[['first_name','last_name','city','country']]
df
```
